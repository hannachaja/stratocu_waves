# Download 4km half-hourly IR grids 

### **MERGEDIR:** 4km, half hourly 
https://disc.gsfc.nasa.gov/datasets/GPM_MERGIR_1/summary?keywords=MERGEDIR. 

OpenDAP filename format: https://disc2.gesdisc.eosdis.nasa.gov/opendap/MERGED_IR/GPM_MERGIR.1/2012/200/merg_2012071800_4km-pixel.nc4.dap.nc4

```python
import numpy as np
import xarray as xr
from datetime import date, timedelta, datetime
import pandas as pd
import matplotlib.pyplot as plt
```

# download MERGE_IR half-hourly images, a centered box

Files are hourly, and contain 2 time levels. 

```python
# Choose the first time, and nhours

pathname = '~/Box/GWaves_MERGIR_data/'
minlat = -40
maxlat = 10
minlon = -130
maxlon = -60
date0 = pd.to_datetime('2024-10-01')
nhours = 24

for i in range(nhours):
    dateGrab = date0 + i*pd.Timedelta(hours=1)
    
# MERGE_IR
    doy = dateGrab.strftime('%Y%j')  # %j gives day of year with zero-padding
    year = dateGrab.strftime('%Y')  # %Y gives year with zero-padding
    mm = dateGrab.strftime('%m')  # %m gives month with zero-padding
    dd = dateGrab.strftime('%d')  # %d gives day with zero-padding
    hh = dateGrab.strftime('%H')  # %H gives hour with zero-padding
        
    filename = f"merg_{year}{mm}{dd}{hh}_4km-pixel.nc4"
    url = f"https://disc2.gesdisc.eosdis.nasa.gov/opendap/MERGED_IR/GPM_MERGIR.1/{year}/{int(doy[4:]):03d}/{filename}"

    print(dateGrab, url)

# sel() the box 
    ds = xr.open_dataset(url)
    LatIndexer, LonIndexer = 'lat', 'lon'
    dsbox = ds.sel(**{LatIndexer: slice(minlat, maxlat),
                      LonIndexer: slice(minlon, maxlon)})
# Write it out
    dsbox.to_netcdf(pathname+filename)   
```
