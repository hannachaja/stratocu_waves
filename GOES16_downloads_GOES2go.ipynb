```python
from pyresample.utils import load_cf_area
from pyresample import create_area_def
from pyresample import kd_tree
from netCDF4 import Dataset
import numpy as np

def resample_goes_cf_to_latlon(netcdf_path, variable_name,
                              target_lat_res=0.1, target_lon_res=0.1,
                              target_extent=(-100, -80, 20, 35)):
    """
    Resample GOES data from a CF-compliant netCDF file to a regular lat-lon grid
    
    Parameters:
    -----------
    netcdf_path : str
        Path to the GOES netCDF file
    variable_name : str
        Name of the variable to resample
    target_lat_res : float
        Resolution of target grid in degrees latitude
    target_lon_res : float
        Resolution of target grid in degrees longitude
    target_extent : tuple
        (min_lon, max_lon, min_lat, max_lat) of target area
    
    Returns:
    --------
    resampled_data : 2D numpy array
        Data resampled to regular lat-lon grid
    lons : 1D numpy array
        Longitude values for the x-axis
    lats : 1D numpy array
        Latitude values for the y-axis (in descending order for correct orientation)
    """
    # Load the source data and its swath definition using CF metadata
    with Dataset(netcdf_path, 'r') as nc:
        source_data = nc.variables[variable_name][:]
        swath_def, cf_info = load_cf_area(netcdf_path)
    
    # Create the target grid definition
    min_lon, max_lon, min_lat, max_lat = target_extent
    
    target_def = create_area_def(
        'target_grid',
        {'proj': 'latlong', 'datum': 'WGS84'},
        area_extent=[min_lon, min_lat, max_lon, max_lat],
        resolution=[target_lon_res, target_lat_res],
        units='degrees'
    )
    
    # Perform the resampling
    resampled_data = kd_tree.resample_nearest(
        swath_def,
        source_data,
        target_def,
        radius_of_influence=50000,  # meters
        fill_value=None
    )
    
    # Create longitude and latitude arrays for plotting
    lons = np.arange(min_lon, max_lon + target_lon_res, target_lon_res)
    lats = np.arange(max_lat, min_lat - target_lat_res, -target_lat_res)  # Descending order
    # lats = np.flip(lats) 
    
    return resampled_data, lons, lats
```

# RT problem for interpreting WV differences 

Claude's response to this prompt: 

I am trying to create a family of atmospheric temperature and humidity profiles in Python, to use as inputs to a radiative transfer model. The central reference sounding should be representative of the Earth's tropics, near a moist adiabat in the troposphere and with a realistic stratosphere on top. It could be a data object, or a computed moist adiabat with a specified relative humidity profile, I don't care which.  For the other soundings, I want profiles representing the crests and troughs of adiabatic gravity waves. These must be computed from the central reference sounding, from sinusiodal vertical displacements of conserved variables in the like specific humidity and moist static energy. I want a family of these, of various vertical wavelengths, with the nodes at 8km altitude, and with the antinodes at 8km altitude. Is this too complicated or can you do it?


```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

class AtmosphericProfiles:
    def __init__(self):
        # Constants
        self.g = 9.81  # m/s^2
        self.Rd = 287.0  # J/kg/K
        self.Rv = 461.5  # J/kg/K
        self.cp = 1004.0  # J/kg/K
        self.L = 2.5e6   # J/kg
        
        # Reference profile setup
        self.z = np.linspace(0, 30000, 301)  # Height in meters
        self.setup_reference_profile()
        
    def setup_reference_profile(self):
        """Create a tropical reference profile"""
        # Surface conditions
        T0 = 300.0  # K
        p0 = 101325.0  # Pa
        RH0 = 0.8  # Surface relative humidity
        
        # Initialize arrays
        self.p = np.zeros_like(self.z)
        self.T = np.zeros_like(self.z)
        self.q = np.zeros_like(self.z)
        
        # Simple tropical temperature profile
        # Troposphere (constant lapse rate to tropopause)
        gamma = 6.5/1000.0  # K/m
        tropopause_height = 16000  # m
        
        for i, z in enumerate(self.z):
            if z <= tropopause_height:
                self.T[i] = T0 - gamma * z
            else:
                # Stratosphere (isothermal)
                self.T[i] = T0 - gamma * tropopause_height
        
        # Hydrostatic pressure
        self.p[0] = p0
        for i in range(1, len(self.z)):
            dz = self.z[i] - self.z[i-1]
            Tbar = 0.5 * (self.T[i] + self.T[i-1])
            self.p[i] = self.p[i-1] * np.exp(-self.g * dz / (self.Rd * Tbar))
        
        # Specific humidity (RH profile decreasing with height)
        self.q = RH0 * self.saturation_mixing_ratio(self.T, self.p) * np.exp(-self.z/7000)
        
        # Calculate moist static energy
        self.h = self.moist_static_energy(self.z, self.T, self.q)
    
    def saturation_vapor_pressure(self, T):
        """Bolton's formula for saturation vapor pressure"""
        return 611.2 * np.exp(17.67 * (T - 273.15) / (T - 29.65))
    
    def saturation_mixing_ratio(self, T, p):
        """Calculate saturation mixing ratio"""
        es = self.saturation_vapor_pressure(T)
        return 0.622 * es / (p - es)
    
    def moist_static_energy(self, z, T, q):
        """Calculate moist static energy"""
        return self.cp * T + self.g * z + self.L * q
    
    def generate_wave_profile(self, wavelength, amplitude, node_at_8km=True):
        """
        Generate a gravity wave profile with specified wavelength and amplitude
        node_at_8km: If True, puts node at 8km; if False, puts antinode at 8km
        """
        # Calculate phase shift to get node or antinode at 8km
        phase_shift = np.pi/2 if node_at_8km else 0
        z_8km = 8000
        
        # Vertical displacement
        dz = amplitude * np.sin(2*np.pi/wavelength * (self.z - z_8km) + phase_shift)
        
        # Interpolate conserved variables to new positions
        z_displaced = self.z - dz
        h_interp = interp1d(self.z, self.h, bounds_error=False, fill_value="extrapolate")
        q_interp = interp1d(self.z, self.q, bounds_error=False, fill_value="extrapolate")
        
        # Get displaced values
        h_new = h_interp(z_displaced)
        q_new = q_interp(z_displaced)
        
        # Solve for temperature (simplified)
        T_new = (h_new - self.g * self.z - self.L * q_new) / self.cp
        
        return T_new, q_new
    
    def plot_profiles(self, wavelengths=[5000, 10000], amplitude=1000, node_at_8km=True):
        """Plot reference and wave profiles"""
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 8))
        
        # Plot reference profile
        ax1.plot(self.T, self.z/1000, 'k-', label='Reference')
        ax2.plot(self.q*1000, self.z/1000, 'k-', label='Reference')
        
        # Plot wave profiles
        colors = plt.cm.rainbow(np.linspace(0, 1, len(wavelengths)))
        for wavelength, color in zip(wavelengths, colors):
            T_wave, q_wave = self.generate_wave_profile(wavelength, amplitude, node_at_8km)
            label = f'λ={wavelength/1000:.1f}km'
            ax1.plot(T_wave, self.z/1000, '--', color=color, label=label)
            ax2.plot(q_wave*1000, self.z/1000, '--', color=color, label=label)
        
        ax1.set_xlabel('Temperature (K)')
        ax1.set_ylabel('Height (km)')
        ax1.grid(True)
        ax1.legend()
        
        ax2.set_xlabel('Specific Humidity (g/kg)')
        ax2.set_ylabel('Height (km)')
        ax2.grid(True)
        ax2.legend()
        
        plt.tight_layout()
        plt.show()

# Example usage
profiles = AtmosphericProfiles()
profiles.plot_profiles(wavelengths=[5000, 10000, 15000], amplitude=2000, node_at_8km=True)
```

```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

class AtmosphericProfiles:
    def __init__(self):
        # Constants
        self.g = 9.81      # m/s^2
        self.Rd = 287.0    # J/kg/K
        self.Rv = 461.5    # J/kg/K
        self.cp = 1004.0   # J/kg/K
        self.cv = 717.0    # J/kg/K
        self.L = 2.5e6     # J/kg
        self.eps = self.Rd/self.Rv
        
        # Reference profile setup
        self.z = np.linspace(0, 30000, 301)  # Height in meters
        self.setup_reference_profile()
    
    def es_bolton(self, T):
        """
        Bolton's formula for saturation vapor pressure (Pa)
        with safety checks for valid temperature range
        """
        T = np.clip(T, 173.15, 373.15)  # Limit to reasonable temperature range
        return 611.2 * np.exp(17.67 * (T - 273.15) / (T - 29.65))
    
    def rs_from_tp(self, T, p):
        """
        Saturation mixing ratio from T and p
        with safety checks for pressure
        """
        es = self.es_bolton(T)
        p = np.maximum(p, es + 1.0)  # Ensure p > es
        return self.eps * es / (p - es)
    
    def dlnesdT(self, T):
        """
        Derivative of ln(es) with respect to T using Bolton's formula
        with temperature range limits
        """
        T = np.clip(T, 173.15, 373.15)
        return 17.67 * 243.5 / ((T - 29.65)**2)
    
    def moist_lapse_rate(self, T, p):
        """
        Calculate moist adiabatic lapse rate (K/m)
        with checks for valid input ranges
        """
        if p < 100 or T < 173.15:  # Switch to dry adiabatic lapse rate for very low pressure/temperature
            return self.g/self.cp
            
        rs = self.rs_from_tp(T, p)
        numerator = self.g * (1 + (self.L * rs)/(self.Rd * T))
        denominator = self.cp + (self.L**2 * rs * self.dlnesdT(T))/(self.Rd * T**2)
        return numerator/np.maximum(denominator, 1e-10)  # Prevent division by zero
    
    def integrate_moist_adiabat(self, T0, p0, RH0):
        """
        Integrate the moist adiabat equations upward
        with improved stability for stratosphere transition
        """
        T = np.zeros_like(self.z)
        p = np.zeros_like(self.z)
        T[0] = T0
        p[0] = p0
        
        tropopause_found = False
        tropopause_index = 0
        
        # Integrate upward using small steps
        for i in range(1, len(self.z)):
            dz = self.z[i] - self.z[i-1]
            
            if not tropopause_found:
                # Fourth-order Runge-Kutta for temperature
                k1 = -self.moist_lapse_rate(T[i-1], p[i-1])
                k2 = -self.moist_lapse_rate(T[i-1] + 0.5*k1*dz, p[i-1])
                k3 = -self.moist_lapse_rate(T[i-1] + 0.5*k2*dz, p[i-1])
                k4 = -self.moist_lapse_rate(T[i-1] + k3*dz, p[i-1])
                T[i] = T[i-1] + (dz/6.0)*(k1 + 2*k2 + 2*k3 + k4)
                
                # Check for tropopause
                if i > 1 and (T[i] > T[i-1] or p[i-1] < 12000):  # ~200 hPa
                    tropopause_found = True
                    tropopause_index = i-1
            
            else:
                # Stratosphere: small positive lapse rate
                T[i] = T[tropopause_index] + 0.002 * (self.z[i] - self.z[tropopause_index])
            
            # Hydrostatic equation for pressure
            Tbar = 0.5 * (T[i] + T[i-1])
            p[i] = p[i-1] * np.exp(-self.g * dz / (self.Rd * Tbar))
        
        return T, p
    
    def setup_reference_profile(self):
        """Create a tropical reference profile using moist adiabat"""
        # Surface conditions
        T0 = 300.0  # K
        p0 = 101325.0  # Pa
        RH0 = 0.8  # Surface relative humidity
        
        # Calculate moist adiabat
        self.T, self.p = self.integrate_moist_adiabat(T0, p0, RH0)
        
        # Calculate humidity profile with better behaved RH profile
        rs = self.rs_from_tp(self.T, self.p)
        RH_profile = RH0 * np.exp(-self.z/7000)  # RH decreasing with height
        RH_profile = np.clip(RH_profile, 0.01, 1.0)  # Ensure RH stays in valid range
        self.q = RH_profile * rs
        
        # Calculate moist static energy
        self.h = self.moist_static_energy(self.z, self.T, self.q)
    
    def moist_static_energy(self, z, T, q):
        """Calculate moist static energy"""
        return self.cp * T + self.g * z + self.L * q
    
    def generate_wave_profile(self, wavelength, amplitude, node_at_8km=True):
        """Generate a gravity wave profile"""
        phase_shift = np.pi/2 if node_at_8km else 0
        z_8km = 8000
        
        dz = amplitude * np.sin(2*np.pi/wavelength * (self.z - z_8km) + phase_shift)
        z_displaced = self.z - dz
        
        h_interp = interp1d(self.z, self.h, bounds_error=False, fill_value="extrapolate")
        q_interp = interp1d(self.z, self.q, bounds_error=False, fill_value="extrapolate")
        
        h_new = h_interp(z_displaced)
        q_new = q_interp(z_displaced)
        
        # Solve for temperature with bounds
        T_new = (h_new - self.g * self.z - self.L * q_new) / self.cp
        T_new = np.clip(T_new, 173.15, 373.15)
        
        return T_new, q_new
    
    def plot_profiles(self, wavelengths=[5000, 10000], amplitude=500, node_at_8km=True):
        """Plot reference and wave profiles"""
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 8))
        
        # Plot reference profile
        ax1.plot(self.T, self.z/1000, 'k-', label='Reference')
        ax2.plot(self.q*1000, self.z/1000, 'k-', label='Reference')
        
        # Plot wave profiles
        colors = plt.cm.rainbow(np.linspace(0, 1, len(wavelengths)))
        for wavelength, color in zip(wavelengths, colors):
            T_wave, q_wave = self.generate_wave_profile(wavelength, amplitude, node_at_8km)
            label = f'λ={wavelength/1000:.1f}km'
            ax1.plot(T_wave, self.z/1000, '--', color=color, label=label)
            ax2.plot(q_wave*1000, self.z/1000, '--', color=color, label=label)
        
        ax1.set_xlabel('Temperature (K)')
        ax1.set_ylabel('Height (km)')
        ax1.grid(True)
        ax1.legend()
        
        ax2.set_xlabel('Specific Humidity (g/kg)')
        ax2.set_ylabel('Height (km)')
        ax2.grid(True)
        ax2.legend()
        
        plt.tight_layout()
        plt.show()

# Example usage
profiles = AtmosphericProfiles()
profiles.plot_profiles(wavelengths=[5000, 10000, 15000], amplitude=2000, node_at_8km=True)
```
